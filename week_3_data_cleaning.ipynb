**week 3 dataset-1**

```python
import pandas as pd
import numpy as np
```

```python
dc = pd.read_csv("/content/Data-cleaning-for-beginners-using-pandas.csv")

dc.reset_index(drop=True,inplace=True)
dc.drop(columns=["Index"],inplace=True)
dc
```

```python
copy_data=dc.copy()
```

```python
#Missing Values
#7 rows in  age column
#1 row in rating column
dc.isnull().sum()
```

```python
#Data Types
dc.info()
```

```python
#converting column names

dc.columns=[x.lower() for x in dc.columns]
dc.columns=[x.replace(" ","_") for x in dc.columns]
dc
```

```python
#filling the missing values in age column with mean value(39)
#mean = 39
dc.age.fillna(dc.age.mean(),inplace=True)

dc["age"] =dc["age"].astype(int)
dc
```

```python
#Salary Formatting
#$ is removed
#salary in thousands(k)
dc.salary=[x.replace("$","") for x in dc.salary]
dc.salary=[x.replace("k","") for x in dc.salary]
dc
```

```python
#rating formatting
#converting nan to nil
#converting -1 to 0

dc.fillna({"rating":"Nil"},inplace=True)
dc["rating"] =dc["rating"].astype(str)

dc.rating=[x.replace("-1","0") for x in dc.rating]
dc
```

dc["location"]=dc["location"].str.replace("India In","India")
dc["location"]=dc["location"].str.replace("IndiaIn","India")
dc["location"]=dc["location"].str.replace("New YorkNy","New York")
dc["location"]=dc["location"].str.replace("Australia Aus","Australia")
dc

```python
#location formatting
dc["location"].str.replace(",","").str.rstrip("In").str.rstrip("Ny").str.rstrip("Aus")
```

```python
#established 	formatting
#replacing -1 with unknown
dc["established"] =dc["established"].astype(str)
dc.established=[x.replace("-1","Unknown") for x in dc.established]
dc
```

```python
#easy_apply formatting
#replace -1 with false

dc["easy_apply"] =dc["easy_apply"].astype(str)
dc.easy_apply=[x.replace("-1","False") for x in dc.easy_apply]
dc.easy_apply=[x.upper() for x in dc.easy_apply]
dc
```

```python
#duplicated entries
ch.duplicated(subset=None, keep=False)
```

```python
ch.choice_description=ch.choice_description.replace(np.nan,"Nil")
ch.head()
```

sample=ch.choice_description.str.split(",",expand=True)
sample

```python
ch.item_price=[x.replace("$","") for x in ch.item_price]
ch.head()
```

```python
#Quantity of each item? (Top 5 itmes available in store)

sample=ch.groupby(["item_name"])["quantity"].sum().sort_values(ascending=False)
print(sample)
print("\n")
sample_1=sample.sort_values(ascending=False).head()
print(sample_1)
```

```python
#which is most ordered item in choice_description?

des=ch.groupby(["choice_description"])["quantity"].sum()
des1=des.sort_values(ascending=False).head(2)

des1.iloc[1:2 :]
```

```python
#What is total order count?

order=ch.groupby(["order_id"]).sum()
order.tail(1)
```

```python
#What is revenue in dollars?

ch["item_price"]=ch["item_price"].astype(float)
ch.item_price.sum()
```

```python
#what are number of orders?

order=ch.groupby(["order_id"]).sum()
order
```

```python
ch.head()
```

```python
#What is average revenue amount per order?

amt=ch.groupby(["order_id"])["item_price"].mean().round(decimals=2)
amt
```

```python
#How many different items are sold?

items=ch.groupby(["item_name"])["quantity"].sum().sort_values(ascending=False)
items.reset_index()
```

```python
#what is revenue from each product?
#in dollars

prod=ch.groupby(["item_name"])["item_price"].sum().sort_values(ascending=False)
prod.reset_index()
```
